##### Channels Notes

* CHANNEL_TYPE contains 8 unique channels (9 if you count nan).
* CHANNEL_ID contains 8 unique ids (9 if you count nan).
* A unique crosswalk between CHANNEL_TYPE and CHANNEL_ID can be created.
* INFO_CHANNEL contains 3 unique options (4 if you count nan) ' active channel and mapped process', ' channel not active', ' active channel and unmapped process'.
* CHANNEL_TYPE_DESC contains 7 unique options
* The CHANNEL_TYPE_DESC is not unique on CHANNEL_TYPE, it contains two options: Branch and Off-site offer
